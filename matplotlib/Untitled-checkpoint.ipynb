```python
x = ['india','america','china','japan']
y = [5,30,19,4]
z= [54,32,45,53]

width = .5
p = np.arange(len(x))
p1 = [j+width for j in p ]

plt.xlabel("country",fontsize =15)
plt.ylabel("gdp",fontsize = 15)
plt.title("world_gdp",fontsize = 15)

c= ["y","b","m","g"]
plt.bar(p,y,width = .4,color = c,align = 'edge',linewidth = 5,edgecolor = 'r',linestyle=':',alpha = .7,label ='world_gdp')
plt.bar(p1,z,width = .4,color = 'r',align = 'edge',linewidth = 5,edgecolor = 'r',linestyle=':',alpha = .7,label ='world_gdp1')

plt.legend()
plt.show()


    
```
